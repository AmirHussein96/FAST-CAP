```python
import os
from pathlib import Path

source_root = Path("/lustre/fsw/portfolios/llmservice/users/amhussein/untarred_data/covost_v2/es-US_en-US")
out_src = Path("/lustre/fsw/portfolios/llmservice/users/amhussein/s2s_latency/source.txt")
out_tgt = Path("/lustre/fsw/portfolios/llmservice/users/amhussein/s2s_latency/target.txt")

def find_audio(root: Path, filename: str):
    """
    Search for `filename` recursively inside 0–39 subfolders.
    Returns file path if found, else None.
    """
    for sub in root.iterdir():
        if sub.is_dir() and sub.name.isdigit():  # folders 0..39
            candidate = sub / filename
            if candidate.exists():
                return str(candidate)
    return None

with out_src.open("w") as src_f, out_tgt.open("w") as tgt_f:
    for cut in cuts_dev:

        # -------------------------------------------------------------
        # 1. Identify the audio filename
        # -------------------------------------------------------------
        audio_name = f"{cut.id}.wav"

        # -------------------------------------------------------------
        # 2. Try to locate audio (search 0–39 folders)
        # -------------------------------------------------------------
        audio_path = find_audio(source_root, audio_name)
        if audio_path is None:
            print(f"WARNING: Audio file not found for cut {cut.id}")
            continue

        # -------------------------------------------------------------
        # 3. Write source path
        # -------------------------------------------------------------
        src_f.write(audio_path + "\n")

        # -------------------------------------------------------------
        # 4. Extract agent-target text
        # -------------------------------------------------------------
        text = " ".join([s.text for s in cut.supervisions if s.speaker == "agent"])
        tgt_f.write(text + "\n")
```

### Combine Cuts to get longform

```python
from collections import defaultdict
import numpy as np
from lhotse import MonoCut, CutSet, Recording, SupervisionSegment
from lhotse.utils import fastcopy
import io
import soundfile as sf

def recording_from_numpy(waveform: np.ndarray, sr: int, rec_id: str = "rec_in_memory"):
    """
    Create a Lhotse Recording from a NumPy waveform entirely in memory.
    waveform shape: (1, num_samples) or (num_samples,) → mono.
    """
    if waveform.ndim == 1:
        waveform = waveform[None, :]  # (1, num_samples)

    # Encode the waveform into WAV bytes in memory
    buffer = io.BytesIO()
    sf.write(buffer, waveform.T, sr, format="WAV")
    wav_bytes = buffer.getvalue()

    # Use Lhotse's from_bytes()
    recording = Recording.from_bytes(data=wav_bytes, recording_id=rec_id)
    return recording

def concat_cuts_per_speaker(
    cuts: CutSet,
    max_duration: float = 30.0,
    speaker_key: str = "origin_spk",
):
    """
    Concatenate cuts belonging to the same speaker until max_duration is reached.

    Args:
        cuts: CutSet of MonoCut
        max_duration: maximum duration (seconds) per concatenated cut
        speaker_key: key in cut.custom identifying the speaker

    Returns:
        CutSet of concatenated MonoCuts
    """

    # 1) group cuts by speaker
    spk2cuts = defaultdict(list)
    for cut in cuts:
        spk = cut.custom.get(speaker_key)
        if spk is None:
            continue
        spk2cuts[spk].append(cut)

    new_cuts = []

    # 2) process each speaker independently
    for spk, spk_cuts in spk2cuts.items():
        
        buf_audio = []
        buf_text = []
        buf_translated_text = []
        buf_supervisions = []
        audio_ids = []
        buf_dur = 0.0
        offset = 0.0
        sr = spk_cuts[0].sampling_rate

        def flush():
            nonlocal buf_audio, buf_supervisions, buf_dur, offset
            if not buf_audio:
                return

            wav = np.concatenate(buf_audio, axis=1)
            id_ = "_".join(audio_ids)
            rec_id = f"{spk}_{len(new_cuts)}"
            recording = recording_from_numpy(wav, sr, rec_id=rec_id)
            text = " ".join(buf_text)
            translated_text = " ".join(buf_translated_text)
            buf_supervisions.append(
                SupervisionSegment(
                    id=id_,
                    recording_id=rec_id,
                    start=0.0,
                    duration=buf_dur,
                    channel=0,
                    text=text,
                    speaker="user",
                )
            )
            buf_supervisions.append(
                SupervisionSegment(
                    id=id_,
                    recording_id=rec_id,
                    start=0.0,
                    duration=buf_dur,
                    channel=0,
                    text=translated_text,
                    speaker="agent",
                )
            )
            new_cut = MonoCut(
                id=id_,
                start=0.0,
                duration=buf_dur,
                channel=0,
                recording=recording,
                supervisions=buf_supervisions,
                custom={"origin_spk": spk},
            )
            new_cuts.append(new_cut)

            buf_audio = []
            buf_supervisions = []
            buf_dur = 0.0
            offset = 0.0

        for cut in spk_cuts:
            if cut.sampling_rate != sr:
                cut = cut.resample(sr)

            if buf_dur + cut.duration > max_duration:
                flush()

            audio = cut.load_audio()  # (1, T)
            text = cut.supervisions[0].text
            translated_text = cut.supervisions[1].text
            buf_audio.append(audio)
            buf_text.append(text)
            buf_translated_text.append(translated_text)
            audio_ids.append(cut.id.split("clips_")[-1])
            offset += cut.duration
            buf_dur += cut.duration

        flush()

    return CutSet.from_cuts(new_cuts)
```

```python
new_cuts_fr = concat_cuts_per_speaker(
    cuts_long_fr,
    max_duration=70.0,
)
new_cuts_es = concat_cuts_per_speaker(
    cuts_long_es,
    max_duration=70.0,
)
new_cuts_de = concat_cuts_per_speaker(
    cuts_long_de,
    max_duration=120.0,
)
```

```python
from re import S
from lhotse import Recording, SupervisionSegment, CutSet, MonoCut
import io
import soundfile as sf
import numpy as np

def recording_from_numpy(waveform: np.ndarray, sr: int, rec_id: str = "rec_in_memory"):
    """
    Create a Lhotse Recording from a NumPy waveform entirely in memory.
    waveform shape: (1, num_samples) or (num_samples,) → mono.
    """
    if waveform.ndim == 1:
        waveform = waveform[None, :]  # (1, num_samples)

    # Encode the waveform into WAV bytes in memory
    buffer = io.BytesIO()
    sf.write(buffer, waveform.T, sr, format="WAV")
    wav_bytes = buffer.getvalue()

    # Use Lhotse's from_bytes()
    recording = Recording.from_bytes(data=wav_bytes, recording_id=rec_id)
    return recording


def combine_cuts(cuts):
    new_cuts = []
    all_cuts = [cut for cut in cuts]
    for i in range(0,len(all_cuts)-1,2):
        supervisions = []
        id_ = f"{all_cuts[i].id}"
        first_start = all_cuts[i].supervisions[0].start
        first_duration = all_cuts[i].duration
        first_text = all_cuts[i].supervisions[0].text
        first_translated_text = all_cuts[i].supervisions[1].text
        second_duration = all_cuts[i+1].duration
        second_text = all_cuts[i+1].supervisions[0].text
        second_translated_text = all_cuts[i+1].supervisions[1].text
        total_duration = first_duration + second_duration
        f_wav = all_cuts[i].load_audio()
        s_wav = all_cuts[i+1].load_audio()

        combined_wav = np.concatenate([f_wav, s_wav], axis=1)
        recording = recording_from_numpy(combined_wav, all_cuts[i].sampling_rate, rec_id=f"{id_}")
        supervisions.append(
            SupervisionSegment(
                id=f"{id_}",
                recording_id=f"{id_}",
                start=first_start,
                duration=total_duration,
                channel=0,
                text=first_text + " " + second_text,
                speaker="user",
            )
        )
        supervisions.append(
            SupervisionSegment(
                id=f"{id_}",
                recording_id=f"{id_}",
                start=first_start,
                duration=total_duration,
                channel=0,
                text=first_translated_text + " " + second_translated_text,
                speaker="agent",
            )
        )
        new_cuts.append(MonoCut(
            id=id_,
            start=first_start,
            duration=total_duration,
            channel=0,
            recording=recording,
            supervisions=supervisions,
        ))
    return CutSet.from_cuts(new_cuts)
```

```python
import os
from pathlib import Path

def find_audio(root: Path, filename: str):
    """
    Search for `filename` recursively inside 0–39 subfolders.
    Returns file path if found, else None.
    """
    for sub in root.iterdir():
        if sub.is_dir() and sub.name.isdigit():  # folders 0..39
            candidate = sub / filename
            if candidate.exists():
                return str(candidate)
    return None

def write_source_target(cuts, output_dir, source_root):
    output_dir.mkdir(parents=True, exist_ok=True)
    out_src = output_dir / "source.txt"
    out_tgt = output_dir / "target.txt"
    with out_src.open("w") as src_f, out_tgt.open("w") as tgt_f:
        for cut in cuts:
            # -------------------------------------------------------------
            # 1. Identify the audio filename
            # -------------------------------------------------------------
            id_ = cut.id.replace("_lustre_fsw_portfolios_llmservice_users_amhussein_data_validation_audio_covost_v2_", "_lustre_fsw_portfolios_edgeai_projects_edgeai_riva_rivamlops_data_AST_data_validation_audio_covost_v2_")
            audio_name = f"{id_}.wav"
            # -------------------------------------------------------------
            # 2. Try to locate audio (search 0–39 folders)
            # -------------------------------------------------------------
            audio_path = find_audio(source_root, audio_name)
            if audio_path is None:
                print(f"WARNING: Audio file not found for cut {cut.id}")
                continue
            # -------------------------------------------------------------
            # 3. Write source path
            # -------------------------------------------------------------
            src_f.write(audio_path + "\n")
            # -------------------------------------------------------------
            # 4. Extract agent-target text
            # -------------------------------------------------------------
            text = " ".join([s.text for s in cut.supervisions if s.speaker == "agent"])
            tgt_f.write(text + "\n")

source_root = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/untarred_data/covost_v2/de-DE_en-US")
output_dir = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/s2s_latency/de-DE_en-US")
write_source_target(shuffled_200_de, output_dir, source_root)

source_root = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/untarred_data/covost_v2/fr-FR_en-US")
output_dir = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/s2s_latency/fr-FR_en-US")
write_source_target(shuffled_200_fr, output_dir, source_root)

source_root = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/untarred_data/covost_v2/es-US_en-US")
output_dir = Path("/lustre/fsw/portfolios/edgeai/users/amhussein/s2s_latency/es-US_en-US")
write_source_target(shuffled_200_es, output_dir, source_root)
```
